
# Masked wordcloud

Using a mask you can generate wordclouds in arbitrary shapes.


```python
import os
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords as nltk_stopwords





# Tlcharge NLTK stopwords dataset si pas dj fait
nltk.download('stopwords')

# Rpertoire de travail courant
d = os.getcwd()

# Chemin vers le rpertoire contenant les fichiers textes
text_dir = path.join(d, '../dumps-text/en/')

# Lis tous les fichiers du rpertoire textes et concatne leur contenu 
all_text = ''
for filename in os.listdir(text_dir):
    if filename.endswith('.txt'):
        with open(path.join(text_dir, filename), 'r') as file:
            all_text += file.read() + ' '

# Ouvre en lecture la mask image sous le bon format
usa_mask = np.array(Image.open(path.join(d, "../images/usa.png")))
# S'assurer que le mask est dans une reprsentation valide : fond en noir et forme en blanc

# Stopwords anglais de NLTK
en_stopwords = set(nltk_stopwords.words('english'))
# Ajout de stopwords personnaliss
en_stopwords.add('J')
en_stopwords.add('p')
en_stopwords.add('e')
en_stopwords.add('g')
en_stopwords.add('one') 
en_stopwords.add('et')  
en_stopwords.add('al')  

# Cre un objet WordCloud en ajustant les options
wc = WordCloud(background_color="white", max_words=100, mask=usa_mask,
               stopwords=en_stopwords, contour_width=1, contour_color='steelblue',
               scale=3)  # Ajuste l'chelle pour une meilleure rsolution

# Gnre le nuage de mot
wc.generate(all_text)

# Sauvegarde le nuage de mot gnr dans un nouveau fichier
wc.to_file(path.join(d, "../images/usa_wordcloud.png"))

# Affiche le nuage de mot
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(usa_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
